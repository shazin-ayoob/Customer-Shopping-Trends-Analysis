```python
import pandas as pd
df = pd.read_csv(r'C:\Analytics study\python d\project 6 amlan\customer_shopping_behavior.csv')
df.head()
```

```python
df.info()
df.describe()
df.describe(include = 'all')
```

```python
df.isnull().sum()
# we can see that there are null values
# its better to substitute with median than mean ,bcoz if mean is used outlier may occur

# also median for each category should be used to give respective rating

df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x : x.fillna(x.median()))
```

```python
df.columns = df.columns.str.lower()
df.columns =df.columns.str.replace(' ','_')
#couldn't rename the below column with above code,so used diff code

df = df.rename(columns={'purchase_amount_(usd)' : 'purchase_amount'})
```

```python
# create a column 'age group'
labels = ['young adult','Adult','Middle-aged','Senior']
df['age_group'] = pd.qcut(df['age'],q=4, labels = labels)
df[['age','age_group']].head(25)
```

```python
df[df['age_group'] == 'Adult']
```

```python
# create a column 'purchase_frequency_day' 

frequency_mapping = {'Fortnightly' : 14, 'Weekly' : 7, 
'Monthly' : 30,'Quarterly' : 90, 'Bi-Weekly' : 14,'Annually' : 365,
'Every 3 Month' : 90}

df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping).astype('Int64')

#used astype(Int64) to avoid decimal in p_f_d

```
